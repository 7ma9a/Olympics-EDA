```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# read the files

athletes = pd.read_csv('athlete_events.csv')
noc = pd.read_csv('noc_regions.csv')
```

```python
athletes.head()
```

```python
noc.head()
```

```python
#Merging the datasets
athletes_df = athletes.merge(noc, how='left',on="NOC")
athletes_df.head()
```

```python
athletes_df.shape
```

```python
athletes_df.rename(columns={'region':'Region','notes':'Notes'}, inplace=True)
athletes_df.head()
```

```python
athletes_df.info()
```

```python
athletes_df.describe()
```

```python
top_10 = athletes_df.Team.value_counts().sort_values(ascending= False).head(10)
top_10
```

```python
plt.figure(figsize=(12,6))
plt.title('Overall Participantion in Olympics by Country')
sns.barplot(x=top_10.index, y= top_10 , palette='Set2')
```

```python
plt.figure(figsize=(12,6))
plt.title('Age distribution of the athletes')
plt.xlabel('Age')
plt.ylabel('Number of participants')
plt.hist(athletes_df.Age, color="orange",bins=np.arange(10,60,2), edgecolor="white")
```

```python
# Total medals

tot_Med = athletes_df.Medal.value_counts()
tot_Med
```

```python
female_participants = athletes_df[(athletes_df.Sex=="F")&(athletes_df.Season=="Summer")][["Sex","Year"]]
female_participants = female_participants.groupby("Year").count().reset_index()
female_participants.tail()
```

```python
women = athletes_df[(athletes_df.Sex=="F")&(athletes_df.Season=="Summer")]
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20,10))
plt.title('Women Participation')
sns.countplot(x="Year", data=women  , palette='Spectral')
```

```python
part = women.groupby('Year')['Sex'].value_counts()
part.head()
```

```python
plt.figure(figsize=(20,10))
part.loc[:,"F"].plot()
plt.title("Plot of female athletes over time")
plt.show()
```

```python
GoldMedals = athletes_df[(athletes_df.Medal=="Gold")].sort_values('Year')
GoldMedals.head()
```

```python
GoldMedals = GoldMedals[np.isfinite(GoldMedals['Age'])]
```

```python
#Golds beyond 60
GoldMedals['ID'][GoldMedals['Age']>60].count()
```

```python
sporting_event = GoldMedals[GoldMedals['Age']>60].Sport
sporting_event
```

```python
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title("Gold Medals for atheltes over 60 years of age")
plt.show()
```

```python
# Gold medal from each country
GoldMedals.Region.value_counts().reset_index(name="Medals").head()
```

```python
totalGoldMedals = GoldMedals.Region.value_counts().reset_index(name='Medal').head(6)
g = sns.catplot(x="index", y="Medal", data=totalGoldMedals, height=5, kind="bar", palette="rocket") 

g.despine(left=True) 
g.set_xlabels("Top 6 countries") 
g.set_ylabels("Number of Medals") 
plt.title('Gold Medals per Country') 
```
